```python
betmgm_home_ml_odds = []
fanduel_home_ml_odds = []

betmgm_away_ml_odds = []
fanduel_away_ml_odds = []

for i in range(len(df)):
    betmgm_home_ml_odds.append(  (df["body"][i])["betmgm"]["homeTeamMLOdds"])
    fanduel_home_ml_odds.append((df["body"][i])["fanduel"]["homeTeamMLOdds"])
    betmgm_away_ml_odds.append(  (df["body"][i])["betmgm"]["awayTeamMLOdds"])
    fanduel_away_ml_odds.append((df["body"][i])["fanduel"]["awayTeamMLOdds"])
    
    

    
def get_rid_of_plus(string):
    if string[0] == "-":
        return float(string)
    else:
        return float(string[1:])

betmgm_home_ml_odds = [get_rid_of_plus(i) for i in betmgm_home_ml_odds]
fanduel_home_ml_odds = [get_rid_of_plus(i) for i in fanduel_home_ml_odds]

betmgm_away_ml_odds = [get_rid_of_plus(i) for i in betmgm_away_ml_odds]
fanduel_away_ml_odds = [get_rid_of_plus(i) for i in fanduel_away_ml_odds]

df["betmgmHomeOdds"] = betmgm_home_ml_odds
df["fanduelHomeOdds"] = fanduel_home_ml_odds

df["betmgmAwayOdds"] = betmgm_away_ml_odds
df["fanduelAwayOdds"] = fanduel_away_ml_odds




#.apply()
```

```python
#step 2: function to calculate surebets
# inputs: two odds from different sportsbooks
#returns: profit percentage, suggested bets for each outcome, returns none if you do not have a surebet

def compute_surebet(odds_a, odds_b):
    implied_prob_a = 1/odds_a
    implied_prob_b = 1/odds_b
    total_implied_prob = implied_prob_a + implied_prob_b
    
    if total_implied_prob <1: #have a surebet opportunity if less than 1
        profit_percentage = (1-total_implied_prob) * 100
        investment = 1000
        
        bet_a = (investment * implied_prob_a) / odds_a
        bet_b = (investment * implied_prob_b) / odds_b
        
        return profit_percentage, bet_a, bet_b
    else: 
        return None, None, None
```

```python

df['arbyHA'] = df.apply(lambda x: compute_surebet(x['betmgmHomeOdds'], x['fanduelAwayOdds']), axis=1)
df['arbyAH'] = df.apply(lambda x: compute_surebet(x['fanduelHomeOdds'], x['betmgmAwayOdds']), axis=1)


#df
#display(df)

ndf = df[["betmgmHomeOdds", "fanduelAwayOdds","fanduelHomeOdds","betmgmAwayOdds","arbyHA","arbyAH"]]
display(ndf)
```
